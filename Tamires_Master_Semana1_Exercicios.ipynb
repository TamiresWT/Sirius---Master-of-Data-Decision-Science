No conjunto de dados `ideb.csv` vamos selecionar as colunas `ano`, `id_municipio`, `media_saeb_mat`, `media_saeb_port` e `media_ideb` provenientes dos dados do Ideb, e as colunas `pib` e `impostos_liquidos` do `pib.csv`.

```python
#criando DataFrame df1 com os dados do ideb
df1 = pd.read_csv('/content/drive/MyDrive/arquivos/ideb.csv')
df1.head()
```

```python
#criando DataFrame df2 com os dados do pib
df2 = pd.read_csv('/content/drive/MyDrive/arquivos/pib.csv')
df2.head()
```

### Atividade 1: Selecione apenas as colunas `ano`, `id_municipio`, `media_saeb_mat`, `media_saeb_port` e `media_ideb` provenientes dos dados do Ideb, apenas para o ano de 2019 e agrupado por município.

```python
df1_filtro = df1.loc[df1['ano'] == 2019, ['ano','id_municipio','nota_saeb_matematica', 'nota_saeb_lingua_portuguesa', 'ideb']]
```

```python
df1_filtro.head()
```

```python
#Agrupando por municipio e incluindo nome de variável agrupada
df1_filtro_agrupado = df1_filtro.groupby(['id_municipio']).mean()
df1_filtro_agrupado.head()
```

```python
#Quantidade de linhas e colunas
df1_filtro_agrupado.shape
```

### Atividade 2: Selecione as colunas `ano`, `id_municipio`, `pib` e `impostos_liquidos` dos dados do PIB, para o ano de 2019 e agrupado por município.

```python
df2_filtro = df2.loc[df2['ano'] == 2019, ['ano','id_municipio','pib', 'impostos_liquidos',]]
df2_filtro.head()
```

```python
#Agrupando por municipio e incluindo nome de variável agrupada
df2_filtro_agrupado = df2_filtro.groupby(['id_municipio']).mean()
df2_filtro_agrupado.head()
```

### Atividade 3: Vamos juntar os dois DataFrames em um único. Para isso, vamos "unir" os dois dataframes pela variável `id_municipio`.

```python
#união de df1_filtro_agrupado com df2_filtro_agrupado a partir da variável id_municipio
df_unico = df1_filtro_agrupado.merge(df2_filtro_agrupado, how = 'outer', on = 'id_municipio')
df_unico.head()
```

### O que significa a correlação das variáveis `id_municipio` e `ideb`?

O Ideb é o Índice de Desenvolvimento da Educação Básica, criado em 2007, que objetiva o monitoramento da qualidade do aprendizado a nível nacional e é utilizado para estabelecer metas para a melhoria do ensino. Dessa forma, a variável ideb aponta a nota do ideb referente a cada município listado no id_municipio. Para chegar a nota, que varia entre zero e dez, o indicador combina informações de desempenho em exames padronizados do Inep (Prova Brasil e Saeb) – obtido pelos estudantes ao final das etapas de ensino (5º e 9º ano do ensino fundamental e 3º ano do ensino médio) – com informações sobre rendimento escolar (aprovação). A forma geral do Ideb é dada por: IDEBji = Nji Pji, em que 0 ≤ Nj ≤10 ; 0 ≤ Pj ≤ 1 e 0 ≤ IDEBj ≤ 10. 

Em que,
i = ano do exame (Saeb e Prova Brasil) e do Censo Escolar;
Nji = média da proficiência em Língua Portuguesa e Matemática, padronizada para um indicador entre 0 e 10, dos alunos da unidade j, obtida em determinada edição do exame realizado ao final da etapa de ensino;
Pji = indicador de rendimento baseado na taxa de aprovação da etapa de ensino dos alunos da unidade j.

Apesar do Ideb ser uma importante ferramenta para acompanhar os avanços e retrocessos referentes à aprendizagem e à aprovação de alunos nas escolas, municípios e estados. Como seu cálculo é definido somente por duas informações, o Ideb tem limitações, que devem ser consideradas num contexto avaliativo, como, por exemplo, vulnerabilidades presentes no território escolar.
